## Heartdisease Data visualization

```python
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(figsize= (10 , 30),subplots = True);
```

## (pyplot(quick) vs matplotlib(advanced))



```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot
over_50.plot(kind='scatter',
            x='age',
            y='chol',
           c='target');
```

```python
#matplot 
import matplotlib.pyplot as plt
fix, ax= plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
           c='target',
             ax = ax );

ax.set_xlim([45,100])
```

```python
## oo method from scratch
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
#Customeize the plot
ax.set(title= "Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

#Add a LEGEND
ax.legend(*scatter.legend_elements(),title = "Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
over_50.head()
```

```python
# Subplot of chol , age , thelach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                             sharex=True)

scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

ax0.set(title="Heart Disease and Cholesterol Levels",
        
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

scatter = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])

ax1.set(title="Heart Disease and thalach Levels",
        xlabel="Age",
        ylabel="thalach")

ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
#Add a title
fig.suptitle("Heart Diease Analysis", fontsize=16 , fontweight ="bold")
```

```python
# Subplot of chol , age , thelach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                             sharex=True)

scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                     cmap="winter")
ax0.set_xlim([50,80])#change the xlim
ax0.set(title="Heart Disease and Cholesterol Levels",
        
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

scatter = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                      cmap="winter")
ax1.set_xlim([50,80])#change the xlim
ax1.set(title="Heart Disease and thalach Levels",
        xlabel="Age",
        ylabel="thalach")

ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
#Add a title
fig.suptitle("Heart Diease Analysis", fontsize=16 , fontweight ="bold")
```
